## For filled csv files, remove the last two columns

```python
import pandas as pd
import os

# ── Input ──────────────────────────────────────────────────────────────────
filled_path = 'csv/CRMLSSold202501_filled.csv' 

# ── Load & Remove Last Two Columns ─────────────────────────────────────────
df = pd.read_csv(filled_path, low_memory=False)

print(f"Original columns ({len(df.columns)}):")
print(list(df.columns))

df = df.iloc[:, :-2]  # drop last two columns

print(f"\nColumns after removing last two ({len(df.columns)}):")
print(list(df.columns))

# ── Save as Non-Filled Version ─────────────────────────────────────────────
output_path = filled_path.replace('_filled', '')
df.to_csv(output_path, index=False)

print(f"\nSaved to: {output_path}")
```

## concat script to merge all months of data

```python
import pandas as pd
import glob
import os

# ── Paths ──────────────────────────────────────────────────────────────────
csv_dir = 'csv'  # adjust if your notebook is not in the root directory

# ── Load & Concat Sold ─────────────────────────────────────────────────────
sold_files = sorted(glob.glob(os.path.join(csv_dir, 'CRMLSSold[0-9]*.csv')))
# This pattern matches CRMLSSold followed by digits only, so _filled files are excluded

print(f"Found {len(sold_files)} sold files:")
for f in sold_files:
    print(f"  {f}")

sold_dfs = []
for f in sold_files:
    df = pd.read_csv(f, low_memory=False)
    print(f"{f}: {len(df)} rows")
    sold_dfs.append(df)

sold = pd.concat(sold_dfs, ignore_index=True)
print(f"\nSold combined row count: {len(sold)}")

# ── Load & Concat Listings ─────────────────────────────────────────────────
listing_files = sorted(glob.glob(os.path.join(csv_dir, 'CRMLSListing[0-9]*.csv')))

print(f"\nFound {len(listing_files)} listing files:")
for f in listing_files:
    print(f"  {f}")

listing_dfs = []
for f in listing_files:
    df = pd.read_csv(f, low_memory=False)
    print(f"{f}: {len(df)} rows")
    listing_dfs.append(df)

listings = pd.concat(listing_dfs, ignore_index=True)
print(f"\nListings combined row count: {len(listings)}")

# ── Filter to Residential Only ─────────────────────────────────────────────
print(f"\nSold property types found: {sold['PropertyType'].unique()}")
print(f"Listings property types found: {listings['PropertyType'].unique()}")

sold_before = len(sold)
listings_before = len(listings)

sold = sold[sold['PropertyType'] == 'Residential']
listings = listings[listings['PropertyType'] == 'Residential']

print(f"\nSold: {sold_before} → {len(sold)} rows after Residential filter (removed {sold_before - len(sold)})")
print(f"Listings: {listings_before} → {len(listings)} rows after Residential filter (removed {listings_before - len(listings)})")

# ── Save ───────────────────────────────────────────────────────────────────
sold.to_csv('sold.csv', index=False)
listings.to_csv('listings.csv', index=False)

print("\nSaved sold.csv and listings.csv to root directory")
```
